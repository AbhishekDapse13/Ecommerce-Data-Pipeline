```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# 1. Generate 5,000 Customers
customers = pd.DataFrame({
    'customer_id': [f'C{i}' for i in range(1, 5001)],
    'name': [f'Customer_{i}' for i in range(1, 5001)],
    'email': [f'user{i}@example.com' for i in range(1, 5001)]
})
customers.to_csv('customers.csv', index=False)

# 2. Generate 50,000 Orders
orders = pd.DataFrame({
    'order_id': [f'O{i}' for i in range(1, 50001)],
    'customer_id': [f'C{random.randint(1, 5000)}' for _ in range(50000)],
    'order_date': pd.date_range(start='2024-01-01', periods=50000, freq='10min') # Changed freq to fit timeframe
})
orders.to_csv('orders.csv', index=False)
```

```python
# 3. Generate Shipments for all 50,000 orders
statuses = ['shipped', 'delivered', 'pending']

# Create random extra hours (0 to 10). 
# Rows with 1-10 extra hours will be > 48 hours total.
extra_hours = [pd.Timedelta(hours=random.randint(0, 10)) for _ in range(50000)]

shipments = pd.DataFrame({
    'shipment_id': [f'S{i}' for i in range(1, 50001)],
    'order_id': orders['order_id'],
    'status': [random.choice(statuses) for _ in range(50000)],
    'shipped_at': orders['order_date'] + pd.Timedelta(days=1),
})

# shipped_at + 2 days (48hrs) + extra hours
shipments['delivered_at'] = shipments['shipped_at'] + pd.Timedelta(days=2) + pd.Series(extra_hours)

shipments.to_csv('shipments.csv', index=False)
```
